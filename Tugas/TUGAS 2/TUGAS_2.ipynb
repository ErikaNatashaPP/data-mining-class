# IMPUTATION MISSING VALUE

```python
np.sum(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
data['Hinselmann'] = data['Hinselmann'].fillna((data['Hinselmann'].median()))
data['Biopsy'] = data['Biopsy'].fillna((data['Biopsy'].median()))
data.isnull().any()
```

```python
# imputasi untuk data numerik yang lain
data['Smokes']=data['Smokes'].fillna(99)
data['IUD']=data['IUD'].fillna(111)
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus
data['Hormonal Contraceptives'] = data.fillna(data['Hormonal Contraceptives'].value_counts().index[0])
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

# ENCODE LABELS

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Type'] = LabelEncoder().fit_transform(data['Smokes'])
data['Hormonal Contraceptives'] = LabelEncoder().fit_transform(data['Hormonal Contraceptives'])
data['Hinselmann'] = LabelEncoder().fit_transform(data['Hinselmann'])
data['Schiller'] = LabelEncoder().fit_transform(data['Schiller'])
data['Citology'] = LabelEncoder().fit_transform(data['Citology'])
data['Biopsy'] = LabelEncoder().fit_transform(data['Biopsy'])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Age'] = StandardScaler().fit(data[['Age']]).transform(data[['Age']])
data['Dx:Cancer'] = StandardScaler().fit(data[['Dx:Cancer']]).transform(data[['Dx:Cancer']])
data['Dx:CIN'] = StandardScaler().fit(data[['Dx:CIN']]).transform(data[['Dx:CIN']])
data['Dx:HPV'] = StandardScaler().fit(data[['Dx:HPV']]).transform(data[['Dx:HPV']])
data['Dx'] = StandardScaler().fit(data[['Dx']]).transform(data[['Dx']])
```

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['Age'] = Normalizer().fit(data[['Age']]).transform(data[['Age']])
data['Dx:Cancer'] = Normalizer().fit(data[['Dx:Cancer']]).transform(data[['Dx:Cancer']])
data['Dx:CIN'] = Normalizer().fit(data[['Dx:CIN']]).transform(data[['Dx:CIN']])
data['Dx:HPV'] = Normalizer().fit(data[['Dx:HPV']]).transform(data[['Dx:HPV']])
data['Dx'] = Normalizer().fit(data[['Dx']]).transform(data[['Dx']])
```
