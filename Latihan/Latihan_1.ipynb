```python
# Multiple key
Biodata = pd.DataFrame({'Nama': ['Erika', 'Erika', 'Irma', 'Irma'],
                        'Travel': ['Jepang', 'Belanda', 'Jepang', 'Bulgaria'],
                        'Status': ['Belum Menikah', 'Belum Menikah', 'Menikah', 'Menikah'], 
                      'Usia': ['21', '20', '25', '26']})

IQ = pd.DataFrame({'Nama': ['Erika', 'Erika', 'Irma', 'Irma'],
                      'Travel': ['Jepang', 'Italia', 'Spanyol', 'Bulgaria'],
                      'IQ': ['117', '198', '200', '127'],
                      'Keterangan': ['Normal', 'Genius', 'Genius', 'Cerdas']})

result3 = pd.merge(Biodata, IQ, on=['Nama', 'Travel'])
result3
```

```python
# Merge methods (default how="inner")
result_left = pd.merge(Biodata, IQ, how='left', on=['Nama', 'Travel'])
result_left
```

```python
# Merge methods (default how="inner")
result_right = pd.merge(Biodata, IQ, how='right', on=['Nama', 'Travel'])
result_right
```

```python
result_outer = pd.merge(Biodata, IQ, how='outer', on=['Nama', 'Travel'])
result_outer
```

```python
result_inner = pd.merge(Biodata, IQ, how='inner', on=['Nama', 'Travel'])
result_inner
```

```python
## Joining 
Biodata = pd.DataFrame({'Status': ['Belum Menikah', 'Menikah', 'Menikah'], 
                      'Usia': ['21', '20', '25']},
                    index=['Erika', 'Irma', 'Niall'])
IQ = pd.DataFrame({'Travel': ['Jepang', 'Italia', 'Spanyol'],
                      'IQ': ['117', '198', '200']},
                    index=['Erika', 'Niall', 'Zayn'])
result = Biodata.join(IQ) #Based Biodata ditambahkan IQ
result
```

```python
result = IQ.join(Biodata, how='outer') #Seluruh data masuk
result
```

```python
result = IQ.join(Biodata, how='inner') # Hanya data yang lengkap
result
```

```python
#The data alignment here is on the indexes (row labels). 
#This same behavior can be achieved using merge plus additional arguments instructing it to use the indexes
result = pd.merge(Biodata, IQ, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(Biodata, IQ, left_index=True, right_index=True, how='inner')
result
```

```python
## Concatenating
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

```python
# Add category
result = pd.concat(frames, keys=['X', 'Y', 'Z'])
result
```

```python
# Default join=outer (join all rows)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

result = pd.concat([df1, df4], axis=0) #axis=1 is for concated by y-axis; default axis=0
result
```

```python
# Join = inner (join only rows that exist in both datasets) 
# hanya yang ada di df1 dan df4
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
# Join the exact rows in df1 
#Menggabungkan berdasarkan df1
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
result
```

```python
## Concatenating using append
result = df1.append(df2) #default axis=0
result
```

```python
result = df1.append(df4)
result
```

```python
result = df1.append([df2, df3]) #multiple datasets to concatenate
result
```

```python
# For DataFrames which dont have a meaningful index, 
#you may wish to append them and ignore the fact that they may have overlapping indexes.
result = pd.concat([df1, df4], ignore_index=True)
result
```
